** How many people have the word Chief in their job title? (This is pretty tricky) **

```python
# to get an idea, we first used a ready-made function
sal[sal['JobTitle'].str.contains('chief', case=False, regex=True)].shape
```

```python
# then, as it was mentioned by the course instructor
# there were some cases in which the word 'chief' was
# followed by comma
def WordChief(title):
    if 'chief' in title.lower().split():
        return True
    elif 'chief,' in title.lower().split():
        return True
    else:
        return False
```

```python
sum(sal['JobTitle'].apply(lambda x: WordChief(x)))
```

** Bonus: Is there a correlation between length of the Job Title string and Salary? **

```python
# create a new column which has shows the length of the Job title for each row
sal['title_len'] = sal['JobTitle'].str.len()
# calculate the correlation for the 2 variables above
sal[['title_len', 'TotalPayBenefits']].corr()
```

There is no correlation between Salary and the length of the Job title, since the correlation coefficient is close to 0.
